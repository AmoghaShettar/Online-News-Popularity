```python
df['shares'].median()
```

```python
Unpop=df[df['shares']<df['shares'].median()]
Pop=df[df['shares']>=df['shares'].median()]
```

```python
data_channel=df.columns[11:17]
data_channel

# Unpop=df[df['shares']<df['shares'].median()]
# Pop=df[df['shares']>=df['shares'].median()]

Unpop_dc = Unpop[data_channel].sum().values
Pop_dc = Pop[data_channel].sum().values

x = np.arange(len(data_channel))
width = 0.35
fig, ax = plt.subplots(figsize=(20,5))
rects1 = ax.bar(x - width/2, Pop_dc, width, label='Popular')
rects2 = ax.bar(x + width/2, Unpop_dc, width, label='Unpopular')

ax.set_ylabel('Count')
ax.set_title('Count of popular and unpoplar news by different new channels')
ax.set_xticks(x,data_channel )
ax.legend()

plt.bar_label(rects1, padding=3)
plt.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
#attributes of news

attr_news= df.columns[[7,8,10]]


# Unpop=df[df['shares']<df['shares'].median()]
# Pop=df[df['shares']>=df['shares'].median()]

Unpop_an = Unpop[attr_news].sum().values
Pop_an = Pop[attr_news].sum().values

x = np.arange(len(attr_news))
width = 0.35
fig, ax = plt.subplots(figsize=(20,5))
rects1 = ax.bar(x - width/2, Pop_an, width, label='Popular')
rects2 = ax.bar(x + width/2, Unpop_an, width, label='Unpopular')

ax.set_ylabel('Count')
ax.set_title('Count of popular and unpoplar news by diffrent attributes in a news (Images, Vidoes, Keywords)')
ax.set_xticks(x,attr_news )
ax.legend()

plt.bar_label(rects1, padding=3)
plt.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
